```python
from datetime import datetime, timedelta

print("=== Calculadora menstrual sencilla ===\n")
print("Esta calculadora solo hace una ESTIMACI√ìN.\n")

# 1. Pedir datos
fecha_str = input("Primer d√≠a de tu √∫ltimo periodo (AAAA-MM-DD): ")
ultima_fecha = datetime.strptime(fecha_str, "%Y-%m-%d").date()

ciclo = int(input("¬øCu√°ntos d√≠as dura tu ciclo? (ej. 28): "))
duracion = int(input("¬øCu√°ntos d√≠as dura tu periodo? (ej. 5): "))

# 2. Calcular pr√≥ximo periodo
proximo_inicio = ultima_fecha + timedelta(days=ciclo)
proximo_fin = proximo_inicio + timedelta(days=duracion - 1)

# 4.D√≠a de ovulaci√≥n aproximado
ovulacion = proximo_inicio - timedelta(days=14)

# 3. Resultados
print("\n=== Resultado ===")
print("√öltimo periodo: ", ultima_fecha)
print("Pr√≥ximo periodo (estimado):")
print("  Empieza: ", proximo_inicio)
print("  Termina:", proximo_fin)
print("\nD√≠a aproximado de ovulaci√≥n:", ovulacion)
print("Solo es una aproximaci√≥n, no es exacto.")
print("\nRecuerda: cada cuerpo es distinto, esto solo es una gu√≠a.")
```

```python
import tkinter as tk
from tkinter import messagebox
from datetime import datetime, timedelta

# L√ìGICA DE LA CALCULADORA

def calcular():
    fecha_str = entry_fecha.get().strip()
    ciclo_str = entry_ciclo.get().strip()
    duracion_str = entry_duracion.get().strip()

    if not fecha_str or not ciclo_str or not duracion_str:
        messagebox.showerror("Error", "Por favor, llena todos los campos.")
        return

    try:
        ultima_fecha = datetime.strptime(fecha_str, "%Y-%m-%d").date()
    except ValueError:
        messagebox.showerror(
            "Error",
            "La fecha debe estar en formato AAAA-MM-DD\nEjemplo: 2025-11-18"
        )
        return

    try:
        ciclo = int(ciclo_str)
        duracion = int(duracion_str)
    except ValueError:
        messagebox.showerror("Error", "El ciclo y la duraci√≥n deben ser n√∫meros enteros.")
        return

    if ciclo <= 0 or duracion <= 0:
        messagebox.showerror("Error", "Los d√≠as deben ser mayores que 0.")
        return

    # C√°lculos
    proximo_inicio = ultima_fecha + timedelta(days=ciclo)
    proximo_fin = proximo_inicio + timedelta(days=duracion - 1)
    ovulacion = proximo_inicio - timedelta(days=14)

    # Mostrar resultados
    label_ultimo_valor.config(text=str(ultima_fecha))
    label_prox_inicio_valor.config(text=str(proximo_inicio))
    label_prox_fin_valor.config(text=str(proximo_fin))
    label_ovulacion_valor.config(text=str(ovulacion))


#INTERFAZ GR√ÅFICA

root = tk.Tk()
root.title("Calculadora Menstrual üíú")

#tama√±o fijo inicial + centrado
window_width = 430
window_height = 580

screen_w = root.winfo_screenwidth()
screen_h = root.winfo_screenheight()

x = (screen_w - window_width) // 2
y = (screen_h - window_height) // 3   # un poco m√°s arriba del centro

root.geometry(f"{window_width}x{window_height}+{x}+{y}")
root.minsize(400, 540)
root.resizable(True, True)

# Colores
fondo = "#fde7ff"
card = "#f7d6ff"
morado = "#a36bff"
morado_oscuro = "#8d55e6"
texto_oscuro = "#3b2245"

root.configure(bg=fondo)

#CONTENEDOR CON SCROLL
container = tk.Frame(root, bg=fondo)
container.pack(fill="both", expand=True)

canvas = tk.Canvas(container, bg=fondo, highlightthickness=0)
scrollbar = tk.Scrollbar(container, orient="vertical", command=canvas.yview)
canvas.configure(yscrollcommand=scrollbar.set)

scrollbar.pack(side="right", fill="y")
canvas.pack(side="left", fill="both", expand=True)

scrollable_frame = tk.Frame(canvas, bg=fondo)

def on_configure(event):
    canvas.configure(scrollregion=canvas.bbox("all"))

scrollable_frame.bind("<Configure>", on_configure)
canvas.create_window((0, 0), window=scrollable_frame, anchor="nw")

#CONTENIDO

titulo = tk.Label(
    scrollable_frame,
    text="Calculadora Menstrual",
    font=("Helvetica", 20, "bold"),
    bg=fondo,
    fg=texto_oscuro
)
titulo.pack(pady=(18, 5))

subtitulo = tk.Label(
    scrollable_frame,
    text="Ingresa tus datos y calcula tu pr√≥ximo periodo\n"
         "y un d√≠a aproximado de ovulaci√≥n",
    font=("Helvetica", 10),
    bg=fondo,
    fg=texto_oscuro
)
subtitulo.pack(pady=(0, 12))

# Tarjeta
frame_card = tk.Frame(scrollable_frame, bg=card, bd=2, relief="ridge")
frame_card.pack(padx=20, pady=10, fill="both")

#Inputs
frame_inputs = tk.Frame(frame_card, bg=card)
frame_inputs.pack(padx=18, pady=18, fill="x")

label_fecha = tk.Label(
    frame_inputs,
    text="Primer d√≠a de tu √∫ltimo periodo\n(AAAA-MM-DD):",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10, "bold")
)
label_fecha.grid(row=0, column=0, sticky="w")

entry_fecha = tk.Entry(frame_inputs, width=18, font=("Helvetica", 10))
entry_fecha.grid(row=0, column=1, padx=(12, 0), pady=5)

label_ciclo = tk.Label(
    frame_inputs,
    text="Duraci√≥n de tu ciclo (d√≠as):",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10, "bold")
)
label_ciclo.grid(row=1, column=0, sticky="w", pady=5)

entry_ciclo = tk.Entry(frame_inputs, width=10, font=("Helvetica", 10))
entry_ciclo.grid(row=1, column=1, padx=(12, 0), pady=5, sticky="w")

label_duracion = tk.Label(
    frame_inputs,
    text="D√≠as que dura tu periodo:",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10, "bold")
)
label_duracion.grid(row=2, column=0, sticky="w", pady=5)

entry_duracion = tk.Entry(frame_inputs, width=10, font=("Helvetica", 10))
entry_duracion.grid(row=2, column=1, padx=(12, 0), pady=5, sticky="w")

#Bot√≥n morado como Label clickable

def btn_enter(event):
    btn_calcular.config(bg=morado_oscuro)

def btn_leave(event):
    btn_calcular.config(bg=morado)

btn_calcular = tk.Label(
    frame_card,
    text="‚ú® Calcular ‚ú®",
    bg=morado,
    fg="white",
    font=("Helvetica", 12, "bold"),
    bd=2,
    relief="raised",
    padx=20,
    pady=8,
    cursor="hand2"
)
btn_calcular.pack(pady=(4, 14))

btn_calcular.bind("<Button-1>", lambda e: calcular())
btn_calcular.bind("<Enter>", btn_enter)
btn_calcular.bind("<Leave>", btn_leave)

#Resultados
frame_resultados = tk.LabelFrame(
    frame_card,
    text=" Resultados ",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10, "bold"),
    labelanchor="n"
)
frame_resultados.pack(padx=18, pady=(4, 16), fill="both")

label_ultimo = tk.Label(
    frame_resultados,
    text="√öltimo periodo:",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10)
)
label_ultimo.grid(row=0, column=0, sticky="w", padx=10, pady=3)

label_ultimo_valor = tk.Label(
    frame_resultados,
    text="-",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10, "bold")
)
label_ultimo_valor.grid(row=0, column=1, sticky="w", padx=10, pady=3)

label_prox_inicio = tk.Label(
    frame_resultados,
    text="Pr√≥ximo periodo (empieza):",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10)
)
label_prox_inicio.grid(row=1, column=0, sticky="w", padx=10, pady=3)

label_prox_inicio_valor = tk.Label(
    frame_resultados,
    text="-",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10, "bold")
)
label_prox_inicio_valor.grid(row=1, column=1, sticky="w", padx=10, pady=3)

label_prox_fin = tk.Label(
    frame_resultados,
    text="Pr√≥ximo periodo (termina):",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10)
)
label_prox_fin.grid(row=2, column=0, sticky="w", padx=10, pady=3)

label_prox_fin_valor = tk.Label(
    frame_resultados,
    text="-",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10, "bold")
)
label_prox_fin_valor.grid(row=2, column=1, sticky="w", padx=10, pady=3)

label_ovulacion = tk.Label(
    frame_resultados,
    text="D√≠a aprox. de ovulaci√≥n:",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10)
)
label_ovulacion.grid(row=3, column=0, sticky="w", padx=10, pady=3)

label_ovulacion_valor = tk.Label(
    frame_resultados,
    text="-",
    bg=card,
    fg=texto_oscuro,
    font=("Helvetica", 10, "bold")
)
label_ovulacion_valor.grid(row=3, column=1, sticky="w", padx=10, pady=3)

# Nota
label_nota = tk.Label(
    scrollable_frame,
    text="‚ö†Ô∏è Esto es solo una estimaci√≥n.\nCada cuerpo es diferente. No sustituye una visita m√©dica.",
    bg=fondo,
    fg=texto_oscuro,
    font=("Helvetica", 9),
    justify="center"
)
label_nota.pack(pady=(6, 14))

root.mainloop()
```
